```python
scan_dom = pd.read_csv('/home/gridsan/lbrenes/hhpred/hhblts/hhmscan_singles_nr/single_gene_systems_NR_hits_dom_nodupes.csv')
```

```python
def gene_size(target_name):
    if target_name == 'mEp332_cIoperonic_homolog_system.blts':
        return 345
    elif target_name == 'mEpX1_Gp15.blts':
        return 62
    elif target_name == 'mEpX1_KilAN.blts':
        return 242
    elif target_name == 'mEpX1_Nun.blts':
        return 110
    elif target_name == 'PhiR12.1_cIoperonic_system.blts':
        return 341
    elif target_name == 'PhiR26.1_SNIPE_system.blts':
        return 500
    elif target_name == 'PhiR46.1_spacklelike.blts':
        return 95
    elif target_name == 'PhiR49.1_KilAN.blts':
        return 165
```

```python
scan_dom['system_len'] = scan_dom['target_name'].apply(gene_size)
```

```python
scan_dom['align_len'] = scan_dom.apply(lambda x: x['to_ali_coord'] - x['from_ali_coord'], axis=1)
scan_dom['align_len_to_tlen'] = scan_dom.apply(lambda x: x['align_len'] / x['tlen'], axis=1)
scan_dom['full_len_comp'] = scan_dom.apply(lambda x: x['qlen'] / x['system_len'], axis=1)
```

```python
scan_dom_filtered=scan_dom[scan_dom["full_len_comp"]<=1.3]
```

```python
scan_dom_filtered=scan_dom_filtered[scan_dom_filtered["full_len_comp"]>=0.7]
```

```python
scan_dom_filtered=scan_dom_filtered[scan_dom_filtered["align_len_to_tlen"]>=0.7]
```

```python
scan_dom_filtered=scan_dom_filtered[scan_dom_filtered["align_len_to_tlen"]<=1.3]
```

```python
scan_dom_filtered.to_csv('single_gene_systems_NR_hits_filtered_dom_nodupes.csv')
```
